```python
from transformers import pipeline
import re
import datetime

llm = pipeline(
    "text2text-generation",
    model="google/flan-t5-large",
    max_length=128
)

print("Agentic AI (type 'exit' to quit)")

def calculator_tool(text):
    """Calculate simple math expressions."""
    try:
        # Extract numbers and operators
        expression = re.findall(r"[0-9]+|[\+\-\*/]", text)
        return str(eval("".join(expression)))
    except:
        return "I couldn't calculate that."

def time_tool():
    """Return current system time."""
    return datetime.datetime.now().strftime("%H:%M:%S")

def chat_tool():
  prompt_chat = f"""
  You are a helpful AI assistant.
  Answer clearly in simple language.
  User: {user_input}
  Assistant:
  """
  return llm(prompt_chat)[0]["generated_text"].strip()

def decide_and_act(user_input):
    # Prompt to decide the tool
    prompt = f"""
You are an AI agent with access to tools: calculator and time checker.
Rules:
1. If the user input contains math, respond exactly: TOOL_CALCULATOR
2. If the user input asks for the current time, respond exactly: TOOL_TIME
3. For all other questions, respond exactly: ANSWER

Do not explain anything. Only output one of these: TOOL_CALCULATOR, TOOL_TIME, ANSWER.

User Input: {user_input}
Agent Decision:
"""
    decision = llm(prompt)[0]["generated_text"].strip()

    print(decision)

    if decision == "TOOL_CALCULATOR":
      return calculator_tool(user_input)
    elif decision == "TOOL_TIME":
      return time_tool()
    else:
      return chat_tool()


while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Agent: Goodbye!")
        break

    result = decide_and_act(user_input)
    print("Agent:", result)
```
